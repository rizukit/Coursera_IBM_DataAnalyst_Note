### Step 1: Import Required Libraries


```python
import pandas as pd
```

### Step 2: Load the Dataset Into a Dataframe


#### **Read Data**
<p>
The functions below will download the dataset into your browser:
</p>


```python
file_path ="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/UDKAZw-kz18Yj8P6icf_qw/survey-data-duplicates.csv"
df = pd.read_csv(file_path)

# Display the first few rows to ensure it loaded correctly
print(df.head())
```

### Step 3. Finding and Removing Duplicates
##### Task 1: Identify duplicate rows in the dataset.


```python
## Write your code here
# 1: Count the number of duplicate rows in the dataset
duplicate_count = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_count}")

# 2: Display the first few duplicate rows to understand their structure
duplicate_rows = df[df.duplicated()]
print("\nFirst few duplicate rows:")
print(duplicate_rows.head())
```

##### Task 2: Remove the duplicate rows from the dataframe.



```python
## Write your code here
# Task 2: Remove the duplicate rows from the dataframe
df=df.drop_duplicates(inplace=False)

# Verification: Count duplicates again to confirm they have been removed
print(f"Number of duplicates remaining: {df.duplicated().sum()}")
```

### Step 4: Finding Missing Values
##### Task 3: Find the missing values for all columns.


```python
## Write your code here
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)
```

```python
## Write your code here
missing_remotework = df['RemoteWork'].isnull().sum()

print(f"Number of missing values in 'RemoteWork': {missing_remotework}")
```

### Step 5. Imputing Missing Values
##### Task 5: Find the value counts for the column RemoteWork.


```python
## Write your code here
remote_work_distribution = df['RemoteWork'].value_counts()

print("Distribution of RemoteWork:")
print(remote_work_distribution)
```

##### Task 6: Identify the most frequent (majority) value in the RemoteWork column.



```python
## Write your code here
# Option 1: Using idxmax() on value_counts (Recommended for clarity)
majority_value = df['RemoteWork'].value_counts().idxmax()

# Option 2: Using the mode() method
# majority_value = df['RemoteWork'].mode()[0]

print(f"The most frequent value in RemoteWork is: {majority_value}")
```

##### Task 7: Impute (replace) all the empty rows in the column RemoteWork with the majority value.



```python
## Write your code here
# Step 1: Identify the majority (mode) value
majority_value = df['RemoteWork'].value_counts().idxmax()

# Step 2: Replace empty rows (NaN) with this majority value
df['RemoteWork'].fillna(majority_value, inplace=True)

# Step 3: Verification - Check if any missing values remain in 'RemoteWork'
remaining_missing = df['RemoteWork'].isnull().sum()
print(f"Majority value used: {majority_value}")
print(f"Missing values remaining in 'RemoteWork': {remaining_missing}")
```

##### Task 8: Check for any compensation-related columns and describe their distribution.



```python
# Task: Describe compensation with safe type-casting to avoid TypeErrors

# 1. Identify columns
comp_cols = df.filter(regex='Comp|Salary').columns

# 2. Generate summary statistics
comp_summary = df[comp_cols].describe()

# 3. FIX: Convert the summary to 'object' type FIRST
# This tells Pandas "I am going to put text/strings in here now"
formatted_summary = comp_summary.astype(object)

# Define metrics that represent money
money_metrics = ['mean', 'std', 'min', '25%', '50%', '75%', 'max']

# 4. Apply formatting safely
# Format ConvertedCompYearly as USD
if 'ConvertedCompYearly' in formatted_summary.columns:
    formatted_summary.loc[money_metrics, 'ConvertedCompYearly'] = \
        comp_summary.loc[money_metrics, 'ConvertedCompYearly'].map('${:,.2f}'.format)

# Format other numeric columns (like CompTotal) as plain decimals
other_cols = [c for c in formatted_summary.columns if c != 'ConvertedCompYearly']
formatted_summary.loc[money_metrics, other_cols] = \
    comp_summary.loc[money_metrics, other_cols].map('{:,.2f}'.format)

# Format 'count' as a plain integer
formatted_summary.loc['count'] = comp_summary.loc['count'].map('{:,.0f}'.format)

print("Compensation Distribution (Formatted Report):")
print(formatted_summary)

#CompTotal is like a bucket filled with $10,000$ Yen, $10,000$ Dollars, and $10,000$ Vietnamese Dong. 
#The "Mean" treats them all as the same "10,000," which is statistically impossible.
```
